```python
import json
from openai import OpenAI

client = OpenAI(
    base_url="https://integrate.api.nvidia.com/v1",
    api_key="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"
)

def get_restaurant_recommendations(
    user_json_path: str,
    restaurant_json_path: str = "/Users/akramchakrouni/Projects/bunq/files/restaurants_dataset.json",
    prompt_path: str = "/Users/akramchakrouni/Projects/bunq/files/system_prompt.txt"
):

    with open(prompt_path, "r", encoding="utf-8") as f:
        system_prompt = f.read()

    with open(restaurant_json_path, "r", encoding="utf-8") as f:
        restaurants_data = json.load(f)

    with open(user_json_path, "r", encoding="utf-8") as f:
        users_data = json.load(f)

    merged_input = json.dumps({
        "restaurants": restaurants_data["restaurants"],
        "user_profiles": users_data["user_profiles"]
    })

    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": merged_input}
    ]

    completion = client.chat.completions.create(
        model="nvidia/llama-3.1-nemotron-70b-instruct",
        messages=messages,
        temperature=0.5,
        top_p=1,
        max_tokens=8096,
        stream=True
    )

    for chunk in completion:
        if chunk.choices[0].delta.content:
            print(chunk.choices[0].delta.content, end="", flush=True)
```
